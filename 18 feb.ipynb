```python
#Ans 1: The API or Application Programming Interface are the software which allows to communication betweeen two different systems.
# for example: The a login page of any website interact with the backend server with the help of APIs.
```

```python
#Ans 2: The advantages of APIs are:
#           - The independence between the systems is maintained.
#           - less resources are required by each system for operating.
#           - can provide customization to the systems.

#       The disadvantages of APIs are:
#           - creating apis is a time consuming process
#           - Costly
```

```python
# Web APIs are the one which uses the HTTP or HTTPS protocol for the transfer of data between the systems. 
# Theses APIs are generally used in web servers and web browsers.
# Difference between API and Web API:-
#    - web apis must require the web servers for the interaction while api can intercet by standard service.
#    - web service use rest or soap architecture while an api can work on any architecture.
#    - web apis are the subset of api.
```
